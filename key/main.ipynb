```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, ttest_ind
import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 11})
pd.set_option('display.max_columns', None)

df = pd.read_csv("bank-additional-full.csv", sep=";")

print("=== ФОРМА ДАННЫХ ===")
print(df.shape)

print("\n=== ПЕРВЫЕ 5 СТРОК ===")
display(df.head())

print("\n=== ТИПЫ ДАННЫХ ===")
print(df.dtypes)

print("\n=== ЗНАЧЕНИЯ 'unknown' В КАТЕГОРИАЛЬНЫХ ПРИЗНАКАХ ===")
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in cat_cols:
    unknown_count = (df[col] == 'unknown').sum()
    if unknown_count > 0:
        print(f"{col}: {unknown_count} ('unknown')")

print("\n=== ПРОПУСКИ (NaN) ===")
print(df.isnull().sum().sum(), "пропущенных значений (ожидаемо 0)")

print("\n=== РАСПРЕДЕЛЕНИЕ ЦЕЛЕВОЙ ПЕРЕМЕННОЙ 'y' ===")
y_counts = df['y'].value_counts()
print(y_counts)
churn_rate = y_counts['yes'] / len(df)
print(f"\n📊 Доля положительных ответов (conversion rate): {churn_rate:.2%}")

# Визуализация целевой переменной
plt.figure(figsize=(6, 4))
y_counts.plot(kind='bar', color=['#FF6B6B', '#4ECDC4'])
plt.title('Целевая переменная: отклик на предложение депозита', fontsize=14, fontweight='bold')
plt.xlabel('Ответ ("no" / "yes")')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. Числовые и категориальные признаки
num_cols = [
    'age', 'duration', 'campaign', 'pdays', 'previous',
    'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed'
]
cat_cols = [col for col in df.columns if col not in num_cols]

# 6. Распределение возраста
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=30, kde=True, color='#4A90E2')
plt.title('Распределение возраста клиентов', fontsize=14, fontweight='bold')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()

# Сравнение возраста по целевой переменной
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='y', y='age', palette=['#FF6B6B', '#4ECDC4'])
plt.title('Возраст: клиенты с откликом vs без', fontsize=14, fontweight='bold')
plt.xlabel('Ответ (y)')
plt.ylabel('Возраст')
plt.tight_layout()
plt.show()

# T-тест для возраста
group_yes = df[df['y'] == 'yes']['age']
group_no = df[df['y'] == 'no']['age']
t_stat, p_val = ttest_ind(group_yes, group_no, equal_var=False)
print(f"\nT-тест для возраста: p-value = {p_val:.3e} → {'Значимо' if p_val < 0.05 else 'Не значимо'}")

# 7. Визуализация категориальных признаков
categorical_for_plot = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

n_cols = 3
n_rows = (len(categorical_for_plot) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(categorical_for_plot):
    if col in df.columns:
        cross_tab = pd.crosstab(df[col], df['y'], normalize='index') * 100
        cross_tab.plot(kind='bar', stacked=True, ax=axes[i], color=['#FF6B6B', '#4ECDC4'])
        axes[i].set_title(f'{col} → доля "yes"')
        axes[i].set_ylabel('%')
        axes[i].tick_params(axis='x', rotation=15)
    else:
        axes[i].axis('off')

# Скрыть лишние subplot'ы
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()

# 8. Анализ маркетинговых переменных
# 8.1 Duration — очень важный признак (продолжительность звонка)
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='duration', hue='y', bins=50, palette=['#FF6B6B', '#4ECDC4'], alpha=0.6)
plt.title('Продолжительность звонка и отклик', fontsize=14, fontweight='bold')
plt.xlabel('Продолжительность звонка (сек)')
plt.ylabel('Частота')
plt.xlim(0, 2000)  # обрезаем хвосты для читаемости
plt.tight_layout()
plt.show()

# 8.2 Campaign — количество контактов в текущей кампании
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='campaign', hue='y', palette=['#FF6B6B', '#4ECDC4'])
plt.title('Количество контактов в текущей кампании и отклик', fontsize=14, fontweight='bold')
plt.xlabel('Номер контакта в кампании')
plt.ylabel('Количество')
plt.xlim(-1, 20)  # фокус на первые 20
plt.legend(title='y', labels=['no', 'yes'])
plt.tight_layout()
plt.show()

# 8.3 Previous и pdays
print("\n=== pdays: 999 означает ранее не контактировали ===")
print(df['pdays'].value_counts().sort_index().head(10))

# Создадим флаг: contacted_before
df['contacted_before'] = df['pdays'].apply(lambda x: 'no' if x == 999 else 'yes')

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='contacted_before', hue='y', palette=['#FF6B6B', '#4ECDC4'])
plt.title('Был ли контакт ранее и отклик сейчас', fontsize=14, fontweight='bold')
plt.xlabel('Ранее контактировали?')
plt.ylabel('Количество')
plt.legend(title='y', labels=['no', 'yes'])
plt.tight_layout()
plt.show()

# 9. Временные паттерны
# 9.1 Месяц
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)

plt.figure(figsize=(10, 5))
monthly = df.groupby('month')['y'].apply(lambda x: (x == 'yes').mean()).reindex(month_order)
sns.barplot(x=monthly.index, y=monthly.values, color='#4ECDC4')
plt.title('Конверсия по месяцам', fontsize=14, fontweight='bold')
plt.xlabel('Месяц')
plt.ylabel('Доля "yes"')
plt.tight_layout()
plt.show()

# 9.2 День недели
day_order = ['mon', 'tue', 'wed', 'thu', 'fri']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=day_order, ordered=True)

plt.figure(figsize=(8, 4))
daily = df.groupby('day_of_week')['y'].apply(lambda x: (x == 'yes').mean())
sns.barplot(x=daily.index, y=daily.values, color='#4ECDC4')
plt.title('Конверсия по дням недели', fontsize=14, fontweight='bold')
plt.xlabel('День недели')
plt.ylabel('Доля "yes"')
plt.tight_layout()
plt.show()

# 10. Экономические индикаторы и занятость
eco_cols = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(eco_cols):
    sns.boxplot(data=df, x='y', y=col, ax=axes[i], palette=['#FF6B6B', '#4ECDC4'])
    axes[i].set_title(f'{col} vs отклик')
    axes[i].set_xlabel('y')

# Скрыть последний пустой subplot
axes[-1].axis('off')

plt.tight_layout()
plt.show()

# 11. Корреляционная матрица (только числовые)
plt.figure(figsize=(12, 8))
corr = df[num_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Корреляции числовых признаков', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 12. Интересные инсайты
print("\n🔍 КЛЮЧЕВЫЕ ИНСАЙТЫ:")
print(f"- Средний возраст: {df['age'].mean():.1f} лет")
print(f"- Медианная длительность звонка: {df['duration'].median()} сек")
print(f"- {((df['pdays'] == 999).mean()*100):.1f}% клиентов не контактировали ранее")
print(f"- Наибольшая конверсия в месяце: {monthly.idxmax().upper()} ({monthly.max():.1%})")
print(f"- Конверсия у клиентов с жилищным кредитом: {(df[df['housing']=='yes']['y']=='yes').mean():.1%}")
print(f"- Конверсия у клиентов без жилищного кредита: {(df[df['housing']=='no']['y']=='yes').mean():.1%}")

# 13. Сохраним очищенный датасет (опционально)
# df_clean = df.replace('unknown', np.nan)
# df_clean.to_csv("bank_clean.csv", index=False)
```

```python
# 1. Импорты
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, mannwhitneyu
import warnings
warnings.filterwarnings('ignore')

# Стиль
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 10})
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 20)

# 2. Загрузка данных
df = pd.read_csv("data.csv")

print("=== ФОРМА ДАННЫХ ===")
print(f"Количество строк: {df.shape[0]}")
print(f"Количество признаков: {df.shape[1] - 1}")
print(f"Целевая переменная: 'Bankrupt?'")

print("\n=== ПЕРВЫЕ 3 СТРОКИ ===")
display(df.head(3))

print("\n=== ТИПЫ ДАННЫХ ===")
print(df.dtypes.value_counts())

# 3. Целевая переменная — дисбаланс?
print("\n=== РАСПРЕДЕЛЕНИЕ ЦЕЛЕВОЙ ПЕРЕМЕННОЙ ===")
target_counts = df['Bankrupt?'].value_counts()
print(target_counts)
print(f"\nДоля банкротств: {target_counts[1] / len(df):.2%}")

# Визуализация дисбаланса
plt.figure(figsize=(6, 4))
target_counts.plot(kind='bar', color=['#4ECDC4', '#FF6B6B'])
plt.title('Распределение целевой переменной: Банкротство', fontsize=14, fontweight='bold')
plt.xlabel('Bankrupt? (0 = нет, 1 = да)')
plt.ylabel('Количество компаний')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 4. Пропуски и дубликаты
print("\n=== ПРОПУСКИ ===")
missing = df.isnull().sum()
print(f"Пропусков: {missing.sum()}")

print("\n=== ДУБЛИКАТЫ ===")
dups = df.duplicated().sum()
print(f"Полных дубликатов: {dups}")

# 5. Описательная статистика
print("\n=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА (числовые признаки) ===")
display(df.describe().T)

# 6. Распределение нескольких ключевых признаков
key_features = [
    'ROA(C) before interest and depreciation before interest',
    'Operating Gross Margin',
    'Net Income to Total Assets',
    'Debt ratio %',
    'Current Ratio',
    'Cash/Total Assets'
]

n = len(key_features)
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, col in enumerate(key_features):
    if col in df.columns:
        sns.histplot(df[col], kde=True, ax=axes[i], color='#4A90E2')
        axes[i].set_title(f'{col}')
        axes[i].set_xlabel('')
    else:
        axes[i].axis('off')

plt.suptitle('Распределение ключевых финансовых показателей', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 7. Сравнение распределений: банкроты vs устойчивые
print("\n=== СРАВНЕНИЕ: БАНКРОТЫ vs НЕ БАНКРОТЫ ===")

# Возьмём топ-10 признаков по абсолютной корреляции с целевой переменной
corr_with_target = df.corr()['Bankrupt?'].abs().sort_values(ascending=False)[1:11]
top_features = corr_with_target.index.tolist()

print("Топ-10 признаков по корреляции с банкротством:")
for i, feat in enumerate(top_features, 1):
    print(f"{i}. {feat}: corr = {corr_with_target[feat]:.3f}")

# Визуализация сравнения
n = len(top_features)
ncols = 3
nrows = (n + ncols - 1) // ncols

fig, axes = plt.subplots(nrows, ncols, figsize=(16, 4 * nrows))
axes = axes.flatten()

for i, col in enumerate(top_features):
    sns.boxplot(data=df, x='Bankrupt?', y=col, ax=axes[i], palette=['#4ECDC4', '#FF6B6B'])
    axes[i].set_title(f'{col}')
    axes[i].set_xlabel('Bankrupt?')

# Скрыть лишние
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle('Сравнение распределений: банкроты (1) vs устойчивые (0)', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 8. Статистическая значимость различий (Mann-Whitney U test — для не нормальных распределений)
print("\n=== СТАТИСТИЧЕСКАЯ ЗНАЧИМОСТЬ РАЗЛИЧИЙ ===")
pvals = {}
for col in top_features:
    group_0 = df[df['Bankrupt?'] == 0][col]
    group_1 = df[df['Bankrupt?'] == 1][col]
    _, p = mannwhitneyu(group_0, group_1, alternative='two-sided')
    pvals[col] = p

sig_features = [f for f, p in pvals.items() if p < 0.05]
print(f"Значимо различаются ({len(sig_features)} из {len(top_features)}):")
for feat in sig_features:
    print(f"  - {feat} (p = {pvals[feat]:.2e})")

# 9. Тепловая карта корреляций (только топ-15 признаков + целевая)
top_corr_cols = ['Bankrupt?'] + corr_with_target.head(15).index.tolist()
plt.figure(figsize=(12, 10))
corr_matrix = df[top_corr_cols].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Корреляции топ-15 признаков с целевой переменной', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 10. Анализ экстремальных значений (выбросы)
print("\n=== ПОТЕНЦИАЛЬНЫЕ ВЫБРОСЫ ===")
# Например: отрицательный долг или ROI > 1000%
extreme_cases = df[
    (df['Debt ratio %'] < 0) |
    (df['ROA(C) before interest and depreciation before interest'] > 10) |
    (df['Net Income to Total Assets'] > 1)
]
print(f"Количество подозрительных записей: {len(extreme_cases)}")

# 11. Инсайты и рекомендации
print("\n" + "="*60)
print("🔍 КЛЮЧЕВЫЕ ИНСАЙТЫ И РЕКОМЕНДАЦИИ")
print("="*60)
print(f"- Датасет сильно несбалансирован: только ~{target_counts[1] / len(df):.1%} банкротств.")
print(f"- Наиболее предиктивные признаки: {', '.join(top_features[:3])}")
print("- Многие финансовые коэффициенты имеют сильные выбросы → рассмотрите winsorizing или лог-преобразование.")
print("- Признаки сильно коррелированы между собой → возможна мультиколлинеарность.")
print("- Для ML рекомендуется: масштабирование, обработка выбросов, балансировка (SMOTE/undersampling).")

# 12. Сохранить очищенный датасет (опционально)
# df.to_csv("data_clean.csv", index=False)
```
